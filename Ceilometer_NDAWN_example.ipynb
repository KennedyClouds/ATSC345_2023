```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import matplotlib.dates as mdates
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

#Read CSV data
csvdata= pd.read_csv('data/sfc/ndawn_20220105.csv',na_values='M')
time=[dt.datetime.strptime(x, '%m/%d/%Y %H:%M') for x in csvdata['time']]
#time19=[dt.datetime.strptime(x, '%Y-%m-%d %H:%M') for x in RDRdata19['valid']]


datafile = 'data/lidar/CL61_010522_5min.nc'
DS = xr.open_dataset(datafile)
data = DS.beta_att.transpose()
data = np.log10(data)
depol = DS.depol.transpose()
#data_rebin = data.coarsen(time=10).mean()

fig=plt.figure(figsize=(6,8),dpi=300)
heights = [1.5, 1.5,1.5,1.5]
spec = fig.add_gridspec(ncols=1, nrows=4,height_ratios=heights)
ax1 = fig.add_subplot(spec[0,0]) 

stuff = data.plot(cmap='Spectral_r',add_colorbar=False,vmin=-8,vmax=-4)
plt.xlim(left=dt.datetime(2022, 1,5, hour=0),right=dt.datetime(2022, 1,5, hour=18))
plt.ylim(0, 750)

ax1.axes.xaxis.set_ticklabels([])
plt.ylabel('Height AGL (m)')
plt.title('CL61 Attenuated Backscatter')
plt.xlabel('')
ax2 = fig.add_subplot(spec[1,0]) 

stuff = depol.plot(cmap='Spectral_r',vmin=0,vmax=1,add_colorbar=False)
plt.xlim(left=dt.datetime(2022, 1,5, hour=0),right=dt.datetime(2022, 1,5, hour=18))
plt.ylim(0, 750)

ax2.axes.xaxis.set_ticklabels([])
plt.ylabel('Height AGL (m)')
plt.title('CL61 Depolarization Ratio')
plt.xlabel('')

ax3 = fig.add_subplot(spec[2,0])
ax33 = ax3.twinx()

ax3.plot([time[0],time[1559]],[15.6464,15.6464],color='grey',linestyle='dashed',linewidth=0.5)
ax3.plot(time,csvdata['windgus'],color='black')
ax3.set_ylim(10,20)
#ax2.plot(time,csvdata['windgus'],color='black',marker='o',color="red")
ax33.plot(time,csvdata['vis']/1000.,color='red')
plt.xlim(left=dt.datetime(2022, 1,5, hour=0),right=dt.datetime(2022, 1,5, hour=18))

plt.ylim(0,5)

ax3.set_ylabel('Wind (m/s)')
ax33.set_ylabel('Visibility (km)', color='red')
plt.xlabel('')
ax3.axes.xaxis.set_ticklabels([])
ax4 = fig.add_subplot(spec[3,0])
ax44 = ax4.twinx()

ax4.plot([time[0],time[1559]],[15.6464,15.6464],color='grey',linestyle='dashed',linewidth=0.5)
ax4.plot(time,csvdata['flowcapt'],color='black')
ax44.plot(time,csvdata['pars'],color='red')
plt.xlim(left=dt.datetime(2022, 1,5, hour=0),right=dt.datetime(2022, 1,5, hour=18))

ax4.set_ylim(0,10)
ax44.set_ylim(0,3000)
ax4.set_ylabel('BLSN Flux $(g/m^2/s)$')
ax44.set_ylabel('PARS Count (#)', color='red')
ax4.set_xlabel('Time (UTC)')
ax4.xaxis.set_major_formatter(mdates.DateFormatter('%H'),)
for tick in ax1.xaxis.get_major_ticks():
    tick.label1.set_horizontalalignment('center')
plt.xticks(rotation=0)


#cbar_ax = fig.add_axes([0.3, 0.02, 0.5, 0.03])
#cbar = fig.colorbar(stuff, cax=cbar_ax,orientation='horizontal')
#cbar.set_label(r'$log_{10}(\beta_{att})$')
plt.savefig('20220105_blsn_CL61.png',dpi=300,bbox_inches='tight')
plt.show()
```
